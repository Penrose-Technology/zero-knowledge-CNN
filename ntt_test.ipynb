```python
import numpy as np
import parameters as p
import ntt_init as n
from sympy import ntt, intt

def ntt_dit(omega_v, x):
    for i in range(p.bitwise):
        m = int(p.N/(2**(i+1)))
        print(m)
        for k in range(0,p.N,2**(i+1)):
            for j in range(i):
                omega = omega_v[j*m]
                u = x[k + j]
                v = x[k + j + 2**i] * omega % p.prime
                x[k + j] = ( u + v ) % p.prime
                x[k + j + 2**i] = ( u - v ) % p.prime
    return x

def bitreverse(s):
    v = np.binary_repr(s, width=p.bitwise)
    v = v[::-1]
    out = int(v, 2)
    return out

def reorder(x):
    x_r = np.zeros_like(x)
    for i in range(p.N):
        x_r[i] = bitreverse(x[i])
    return x_r
```

```python
x = np.array([int(i) for i in range(p.N)])
x_r = reorder(x)
print(x)
print(x_r)
a_ref = np.array(ntt(x, prime=p.prime))
print(intt(a_ref, p.prime))

omega = n.calculate_omega()
print(f"polynomial degree is {p.N}, primitive root omega is {omega}")
A = n.N_roots_of_unit(omega)
print(f"ntt roots of unit vector is \n{A}")
a = ntt_dit(A, x_r)
print(a)
```
